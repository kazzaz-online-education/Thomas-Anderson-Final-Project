```python
queryYearlyClose = """
select 
  substr(date, 1,4) as year, 
  max(close) as max_close, 
  avg(close) as avg_close, 
  min(close) as min_close
from sp500
group by substr(date, 1,4)
order by substr(date, 1,4) desc
limit 20
"""
#substr extracts lengths of characters from a string. In this case, we are separating each set out by the year. Then choosing the parameters from there.
# With "desc" at the end of the "order by", it takes the most recent 20 years. If we did not put that in, it would take the first 20 years. 
resultYearlyClose = pd.read_sql_query(queryYearlyClose, dbSp500)

print("SQL Result: \n\n %s" % resultYearlyClose)
```
